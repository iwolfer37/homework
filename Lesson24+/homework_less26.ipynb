```python
# 2 завданя
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# вибірка
data = np.array([10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7 ,9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12])

counter = Counter(data)
```

```python
# Сортуємо
values = np.array(sorted(list(counter.keys())))

# Розрахунок частот
frequencies = np.array([counter[val] for val in values])

# Розрахунок емпіричної функції
cumulative_frequencies = np.cumsum(frequencies) / len(data)
```

```python

# Полігон частот (дяка чату гпт)
plt.figure(figsize=(10, 5))
plt.stem(values, frequencies, use_line_collection=True)
plt.title('Полігон частот')
plt.xlabel('Значення')
plt.ylabel('Частота')
plt.show()

# Графік емпіричної функції розподілу
plt.figure(figsize=(10, 5))
plt.step(values, cumulative_frequencies)
plt.title('Емпірична функція розподілу F*(x)')
plt.xlabel('Значення')
plt.ylabel('F*(x)')
plt.show()

# Розрахунок моди
mode = values[np.argmax(frequencies)]
print(f'Мода вибірки: {mode}')

# Розрахунок медіани
median = np.median(data)
print(f'Медіана вибірки: {median}')

```
