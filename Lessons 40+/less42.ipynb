```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Завантажуємо винний датасет й нормалізуємо дані
wine = load_wine()
wine_data = wine.data

scaler = StandardScaler()
wine_data = scaler.fit_transform(wine_data)
```

```python
# переводим в 2D для візуалізації
pca = PCA(n_components=2)
wine_data = pca.fit_transform(wine_data)

# KMeans кластеризація
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(wine_data)

# DBSCAN кластеризація
dbscan = DBSCAN(eps=1)
dbscan_clusters = dbscan.fit_predict(wine_data)

# Ієрархічна кластеризація
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_clusters = hierarchical.fit_predict(wine_data)
```

```python
# Візуалізація результатів
fig, axs = plt.subplots(3, figsize=(15,15))

axs[0].scatter(wine_data[:, 0], wine_data[:, 1], c=kmeans_clusters)
axs[0].set_title('KMeans Clustering')

axs[1].scatter(wine_data[:, 0], wine_data[:, 1], c=dbscan_clusters)
axs[1].set_title('DBSCAN Clustering')

axs[2].scatter(wine_data[:, 0], wine_data[:, 1], c=hierarchical_clusters)
axs[2].set_title('Hierarchical Clustering')

for ax in axs.flat:
    ax.set(xlabel='PCA1', ylabel='PCA2')

plt.show()
```
