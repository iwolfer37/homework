```python
# Завдання 2 (в першому я в excel створив табличку з авто)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Завантаження даних з Excel
df = pd.read_excel('car.xls')
```

```python
# Графік: споживання палива від вартості
plt.plot(df['Приблизна вартість, $'], df['Росхід палива, л'], 'o-')
plt.xlabel('Вартість, $')
plt.ylabel('Споживання палива, л')
plt.title('Залежність споживання палива від вартості')
plt.grid(True)
plt.show()

# Графік: споживання палива від швидкості
plt.scatter(df['Максимальна швидкість, км/год'], df['Росхід палива, л'])
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Споживання палива, л')
plt.title('Залежність споживання палива від швидкості')
plt.grid(True)
plt.show()

# Графік: вантажопідйомність від швидкості
plt.scatter(df['Максимальна швидкість, км/год'], df['Вантажопідйомність, кг'])
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Вантажопідйомність, кг')
plt.title('Залежність вантажопідйомності від швидкості')
plt.grid(True)
plt.show()

```

```python
# Завдання 3. Розділення автомобілів на 5 груп за швидкістю
speed_group = 5
df['Швидкість група'] = pd.cut(df['Максимальна швидкість, км/год'], bins=speed_group)

# Перетворення значень інтервалів на рядки
df['Швидкість група'] = df['Швидкість група'].astype(str)

# Побудова гістограми
plt.hist(df['Швидкість група'], bins=speed_group)
plt.xlabel('Швидкість група')
plt.ylabel('Кількість автомобілів')
plt.title('Розподіл автомобілів за швидкістю')
plt.xticks(rotation=45)
plt.show()
```

```python
# Групування даних за марками автомобілів
grouped_df = df.groupby('Марка й модель')

# Побудова коробкових графіків для кожної марки автомобіля
plt.figure(figsize=(12, 6))
positions = np.arange(len(grouped_df))
plt.boxplot([group_data['Максимальна швидкість, км/год'] for group_name, group_data in grouped_df], positions=positions)
plt.xlabel('Марка автомобіля')
plt.ylabel('Швидкість, км/год')
plt.title('Розподіл швидкостей різних марок автомобілів')
plt.xticks(positions, [group_name for group_name, _ in grouped_df], rotation=90)
plt.show()
```
