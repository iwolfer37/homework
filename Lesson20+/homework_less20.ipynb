```python

import random
import string

int_list = []
float_list = []
str_list = []

for i in range(5000):
    int_list.append(random.randint(0, 1000))

for i in range(5000):
    float_list.append(round(random.uniform(0.1, 100.0), 2))

for i in range(5000):
    str_list.append(''.join(random.choices(string.ascii_lowercase, k=random.randint(5, 10))))

print("int_list:", int_list)
print("float_list:", float_list)
print("str_list:", str_list)
```

```python
#  2
import random
import string
import time

#      
def sort_time_sered(lst, num_iters, sort_func):
    total_time = 0
    for i in range(num_iters):
        start_time = time.time()
        sort_func(lst)
        total_time += time.time() - start_time
    return total_time / num_iters

#   int_list
int_list_sorted = sorted(int_list)

#   float_list
float_list_sorted = sorted(float_list)

#   str_list
str_list_sorted = sorted(str_list)

#      
def my_sort(lst):
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            if lst[j] < lst[i]:
                lst[i], lst[j] = lst[j], lst[i]

my_sort(int_list)
int_list_sorted_iwolfer = int_list

my_sort(float_list)
float_list_sorted_iwolfer = float_list

my_sort(str_list)
str_list_sorted_iwolfer = str_list

print("int_list_sorted:", int_list_sorted)
print("float_list_sorted:", float_list_sorted)
print("str_list_sorted:", str_list_sorted)

print("int_list_sorted_iwolfer:", int_list_sorted_iwolfer)
print("float_list_sorted_iwolfer:", float_list_sorted_iwolfer)
print("str_list_sorted_iwolfer:", str_list_sorted_iwolfer)

num_iters = 10
print("    int_list_sorted_iwolfer:", sort_time_sered(int_list, num_iters, my_sort))
print("    float_list_sorted_iwolfer:", sort_time_sered(float_list, num_iters, my_sort))
print("    str_list_sorted_iwolfer", sort_time_sered(str_list, num_iters, my_sort))
```

```python
%%timeit

num_iters = 10
sort_time_sered(int_list, num_iters, my_sort)
```
