```python
import cv2
import matplotlib.pyplot as plt

# Зчитування оригінального зображення
image_path = '/Users/iWolfer/Desktop/Pycharm/homework(pds6)/Lessons 50+/59/my_photo.jpg'
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('моя фотка оригінал')
plt.axis('off')
plt.show()
```

```python
# Фільтр Собеля
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)

sobel_result = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
plt.imshow(cv2.cvtColor(sobel_result, cv2.COLOR_BGR2RGB))
plt.title('Фільтр Собеля')
plt.axis('off')
plt.show()
```

```python
# Медіанний фільтр
median_filtered = cv2.medianBlur(image, 5)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Медіанний фільтр')
plt.axis('off')
plt.show()
```
