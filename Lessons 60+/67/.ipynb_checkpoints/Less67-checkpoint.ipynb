```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Створюємо випадковий датасет
np.random.seed(0)
X = np.linspace(0, 10, 100)  # Згенеруємо 100 точок в діапазоні від 0 до 10
y = 2 * X + 1 + np.random.normal(0, 1, 100)  # Лінійна залежність з шумом
```

```python
# Побудова графіка
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Графік згенерованого датасету')
plt.show()

# Розділення даних на тренувальний і тестовий набори
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# Створення моделі лінійної регресії за допомогою TensorFlow
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])
```

```python
# Компіляція моделі
model.compile(optimizer='sgd', loss='mean_squared_error')

# Навчання моделі на тренувальних даних
history = model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
# Оцінка моделі на тестових даних
loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Втрати на тестових даних: {loss}')
```

```python
# Отримання прогнозів моделі для всього діапазону X
y_pred = model.predict(X)
```

```python
# Візуалізація результатів
plt.scatter(X, y, label='Дані')
plt.plot(X, y_pred, color='red', linewidth=3, label='Лінійна регресія')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Результати лінійної регресії')
plt.show()
```
