```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Зчитування оригінального зображення
image_path = '/Users/iWolfer/Desktop/Pycharm/homework(pds6)/Lessons 60+/60/my_cat_tessi.jpg'
image = cv2.imread(image_path)

# Конвертація зображення в чорно-біле
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Застосування фільтра Собеля для вертикальних та горизонтальних границь
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)

# Відображення результатів
plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Оригінал')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Застосування фільтру Canny для виділення границь
canny_image = cv2.Canny(gray_image, threshold1=30, threshold2=100)

# Відображення результату
plt.figure(figsize=(8, 6))
plt.imshow(canny_image, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

```python
# Визначення фільтра Робертса
roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])

# Застосування фільтра Робертса для виділення границь
roberts_x_image = cv2.filter2D(gray_image, cv2.CV_64F, roberts_x)
roberts_y_image = cv2.filter2D(gray_image, cv2.CV_64F, roberts_y)

# Об'єднання результатів по модулю для отримання фінального зображення
roberts_edge_image = np.abs(roberts_x_image) + np.abs(roberts_y_image)

# Відображення результату
plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 2)
plt.imshow(np.abs(roberts_x_image), cmap='gray')
plt.title('Roberts X')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(np.abs(roberts_y_image), cmap='gray')
plt.title('Roberts Y')
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.imshow(roberts_edge_image, cmap='gray')
plt.title('Roberts Edge Detection')
plt.axis('off')
plt.show()
```
